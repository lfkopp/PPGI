```python
import pandas as pd
import numpy as np

dir = 'C:/_dados/'
```

```python
train = pd.read_csv(dir+'consolidated_text.txt', sep=';',index_col=0,names=['txt'])
c = pd.read_pickle('DONORSCHOOSE\words.pickle')
train.head()
```

```python
words = c.index.tolist()
words[:10]
```

```python
df = pd.DataFrame()
train=train.head(16000)
for word in words:
    df[word] = train.txt.apply(lambda x: (word in x)*1)
df.to_pickle('DONORSCHOOSE\matrix.pickle')
```
