```python
import pandas as pd
import numpy as np
from nltk.stem import SnowballStemmer

dir = 'C:/_dados/_donors/'
```

```python
train = pd.read_csv(dir+'consolidated_text.txt', sep=';',index_col=0,names=['id','approved','category','sentence'])
stemmer = SnowballStemmer("english")
```

```python
train.head()
```

```python
groups = []
for x in train.groupby('category')['category'].head(1):
    if x not in groups: groups.append(x)
groups
```

```python
def convert(frase):
    frase2 = []
    for word in frase.split(' '):
        frase2.append(stemmer.stem(word))
    return ' '.join(frase2)

for group in groups:
    print(group)
    train2 = train[train.category == group]
    train2['sentence'] = train2['sentence'].apply(convert)
    display(train2.head())
    train2.to_pickle(dir+'/matrix_cat_'+group+'.pickle')
    
```

```python

for group in groups:
    df = pd.read_pickle(dir +'categories/matrix_cat_'+ group +'.pickle')
    for ap in [0,1]:
        df2=df[df['approved'] == ap]
        c = df2['sentence'].str.split(' ')
        for line in c:
            print(line)
        df2.to_pickle(dir + 'matrix/matrix_' + group + '_'+ str(ap) + '.pickle')
    break
```
