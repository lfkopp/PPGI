```python
import pandas as pd
import numpy as np
```

```python
dir = 'C:/_dados/_avito/'
pd.options.display.max_rows = 99
train = pd.read_csv(dir + 'train.csv').fillna(0)
test = pd.read_csv(dir + 'test.csv').fillna(0)
```

```python
train.head()
```

```python
test.head()
```

```python
train['source'] = 'train'
test['source'] = 'test'
df = train.append(test, ignore_index=True)
display(df.head())
display(df.shape)
```

```python
del train
del test
```

```python
colunas = []
remover_col = []

df['user_cat'] = df['user_type'] + df['parent_category_name']

for i,col in enumerate(df.columns):
    tam = len(df[col].unique())
    print(i, col, tam)
    if(2<tam<4000): 
        colunas.append(col)
    else:
        remover_col.append(col)
display(colunas)
display(remover_col)
```

```python
colunas.remove('activation_date')
colunas.remove('city')
colunas.remove('param_3')
colunas.remove('param_2')

remover_col.remove('deal_probability')
remover_col.remove('source')
remover_col.remove('item_id')

print(colunas)
print(remover_col)
```

```python
chute1 = pd.DataFrame(df[df['source'] == 'train'].groupby(['category_name', 'user_type'])['deal_probability'].mean())
chute1.reset_index(inplace=True)
chute1.head()
```

```python
chute1.rename({'deal_probability':'deal_chute1'}, axis='columns', inplace=True)
chute1.head()
```

```python
df2 = pd.merge(df, chute1, on=['category_name', 'user_type'], how='left')
df2.drop(remover_col, axis=1, inplace=True)
df2.head()
```

```python
co = ['activation_date','city','param_2','param_3']
df2.drop(co, axis=1, inplace=True)
df2.head()
```

```python
for i,col in enumerate(colunas):
    for j,group in enumerate(df2[col].unique()):
        if (j > 0) and (df2[df2[col] == group]['item_id'].count() > 5000):
            df2[str('col') + '_' + str(i) + '_' + str(j)] = (df2[col] == group) * 1
df2.head()
```

```python
df2.drop(colunas, axis=1, inplace=True)
df2.head()
```

```python
del df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
X_train = df2[(df2['source'] == 'train')]
X_test  = df2[(df2['source'] == 'test')]
```

```python
from sklearn.linear_model import LinearRegression
model =  LinearRegression()

predictors = X_train.iloc[:,3:].values * 1
target     = X_train['deal_probability'].values
test       = X_test.iloc[:,3:].values * 1
label      = X_test['item_id'].values

model.fit(predictors, target)
y_pred = model.predict(test)
print(y_pred[:20])
```

```python
with open(dir + 'submissions/basicao_003.csv', 'w') as f:
    f.write('item_id,deal_probability\n')
    for label_, predict_ in zip(label, y_pred):
        p = str(min(1,max(0,float(predict_))))
        f.write(str(label_) + ',' + p + '\n')
        
#### SCORE 0.2415
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(max_depth=2, random_state=0)
predictors = X_train.iloc[:,3:].values * 1
target     = X_train['deal_probability'].values
test       = X_test.iloc[:,3:].values * 1
label      = X_test['item_id'].values

model.fit(predictors, target)
y_pred = model.predict(test)
print(y_pred[:20])
```

```python
with open(dir + 'submissions/basicao_004.csv', 'w') as f:
    f.write('item_id,deal_probability\n')
    for label_, predict_ in zip(label, y_pred):
        p = str(min(1,max(0,float(predict_))))
        f.write(str(label_) + ',' + p + '\n')
        
#### SCORE 0.2415
```
