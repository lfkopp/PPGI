```python
import pandas as pd
import numpy as np
import seaborn as sns
import re
from collections import Counter
from nltk import download
from nltk import word_tokenize
from nltk.stem import SnowballStemmer
from nltk.corpus import stopwords
```

```python
%matplotlib inline
dir = 'c:/_dados/_avito/'
download('punkt')
stemmer = SnowballStemmer("russian")
stop = set(stopwords.words('russian'))
```

```python
def convert(frase):
    frase2 = []
    for word in str(frase).split():
        frase2.append(stemmer.stem(word))
    return ' '.join(frase2)
```

```python
groups = []
for x in df.groupby('parent_category_name')['parent_category_name'].head(1):
    if x not in groups: groups.append(x)
groups
```

```python
df2 = df[['item_id', 'parent_category_name', 'source', 'description']]
i =0
for group in groups:
    i += 1
    df5 = df2[df2['parent_category_name'] == group]
    df4 = df5[df5['source'] == 'train'] #[:5000]
    with open(dir+'consolidated_'+str(i)+'.txt', 'wb') as f: 
        for idx, row in df4.iterrows():
            idx = row.item_id
            r = re.sub("[\(\);:,.*\+\_\-\%!?a-zA-Z0-9“”\\\\/=«»—\"\'®•💋↓·&😍😏👏🌴🔥🍓🚩👰💃🎂✔🎁⇘⇓⇙◄►▬❀✈🎄🎉🎤☀🌍⇒₽~]", " ", str(row.description))
            r = r.replace('nan', ' ')
            r = re.sub("\s\s+", " ", r)
            r = r.split()
            r = [x.lower() for x in r if len(x)>2]
            r = [x for x in r if x not in stop]
            con = Counter(r).items()
            content = ''
            for item in con:
                a,b = item
                content += str(group) + ';' +str(idx) + ';' + stemmer.stem(str(a)) + ';' + str(b) + "\r\n"
            f.write(content.encode('utf-8'))
    df5 = pd.read_csv(dir + 'consolidated_'+str(i)+'.txt', sep=';', names=['parent_category', 'item_id', 'word_stem', 'number'])
    #display(stemmed.head())
    stemmed = df5.groupby(['item_id','word_stem'],as_index = False)['number'].agg('sum').pivot('item_id','word_stem').fillna(0)
    stemmed.to_pickle(dir + 'stemmed_' + str(i) + '.pickle')
```
