```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn import svm
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from IPython.display import clear_output
```

Após importar as bibliotecas, carrega-se a base de dados, já disponível no sklearn. A Descrição da base é apresentada em seguida.

```python
iris = load_iris()
X = iris.data
y = iris.target
print(iris.DESCR)
```

```python
for x in [0,50,100]:
    display(pd.DataFrame(X[x:x+50]).describe())
```

```python
class Perceptron:

    def __init__(self, learning_rate = .1):
        self.w = None
        self.learning_rate = learning_rate
        self.input_size = 0
        self.output_size = 0
        
    def fit(self, X, y, n_iter=1000):
        self.sample_size = X.shape[1]
        self.input_size = len(y)
        self.output_size = len(np.unique(y))
        self.w = np.random.rand(self.output_size, self.sample_size)
        self.b = np.random.rand(self.output_size)
        
        y2 = np.zeros((self.input_size, self.output_size))
        for u,t in enumerate(y):
            y2[u][t] = int(1)

            
        for i in range(n_iter):
            for posi in range(self.sample_size):
                posi = np.random.randint(0,X.shape[0])
                x = X[posi]
                y_expected = y2[posi]

                Z = np.dot(self.w,x) + self.b

                y_hat = self._sigmoid(Z)
                error = (y_hat - y_expected)

                for j in range(self.output_size):
                    gradient = error[j] * self._sigmoid(Z)[j]
                    self.w[j] -= self.learning_rate * gradient
                    self.b -= self.learning_rate * error[j] * self._sigmoid(Z)[j]
        return self

    def predict(self, X):
        Z = np.array([np.dot(self.w,x) + self.b for x in X])
        return np.argmax(self._sigmoid(Z), axis=1)

   

    def _sigmoid(self, z): 
        return 1/(1 + np.exp(-z))
    
    def _sigmoid_dev(self, z):
        return self._sigmoid(z) * (1-self._sigmoid(z))

    def _acc(self, X, y):
        count = 0
        y_hat = self.predict(X)
        for i in range(len(y)):
            if np.argmax(y[i]) == y_hat[i]:
                count += 1

        return count / len(y) * 1.0
```

```python
np.random.seed(40)
acc_list = []
for i in range(30):
    X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.3)

    #criando o classificador
    c = Perceptron(learning_rate=0.2)

    # vamos aprender!
    c.fit(X_train, y_train, n_iter=5000)

    # utilizando o modelo para prever
    y_hat = c.predict(X_test)
    # medindo acertos
    acc_list.append(sum(y_hat == y_test)/len(y_test))
    #clear_output()
    print("ACC MEAN: ", np.round(np.mean(acc_list),5), "ACC STD: ", np.round(np.std(acc_list),5),flush=True)
    
```

Abaixo, aplicamos a classe perceptron nos dados diversas vezes, para obter significância estatística nos resultados.

Com 2000-2500 iterações já temos resultados acima de 90% de acerto.

```python
plt.hist(acc_list,bins=20,range=(0,1))
plt.show()
```

### O benchmarking é a função otimizada da própria biblioteca do sklearn

```python
acc_list = []
np.random.seed(42)
for i in range(100):
    X_train, X_test, y_train, y_test  = train_test_split(X, y, test_size=0.33)
    
    #criando o classificador
    clf = svm.SVC()
    
    # vamos aprender!
    clf.fit(X_train, y_train)
    
    # utilizando o modelo para prever
    y_hat = clf.predict(X_test)

    
    acc = accuracy_score(y_test, y_hat)
    acc_list.append(acc)

print("NN do sklearn")
print("ACC MEAN: ", np.mean(acc_list))
print("ACC STD: ", np.std(acc_list))
```

```python
plt.hist(acc_list,bins=20,range=(0.6,1))
plt.show()
```
