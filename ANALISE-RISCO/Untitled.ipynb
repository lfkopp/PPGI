```r
#1-Simule um jogo de cara ou coroa. Verifique a frequencia do nÃºmero de caras com: 10, 100 e 1000 lanÃ§amentos.
ex1 = function () {
	for(N in c(10,100,1000))
	print(sum(sample(0:1,N,replace=T)))
}
ex1()
```

```r
#2- Seja X uma VA que representa a soma de 12 VAs uniformes (0,1):
#2a- Usando o TCL, obtenha uma aproximaÃ§Ã£o analÃ­tica para a X, e calcule a mÃ©dia, a variÃ¢ncia e esboce um grÃ¡fico dessa distribuiÃ§Ã£o.
ex2a <- function() {
  
  total.samples <- switch("U", "U" = matrix(runif(12000, min = 0, max = 1),120))
  
  for(i in 1:11) { 
    my.samples <- switch("U", "U" = matrix(runif(12000, min = 0, max = 1),120))
    total.samples <- rbind(my.samples,total.samples)
  }
  print(total.samples)
  all.sample.sums <- apply(total.samples,1,sum)
  all.sample.means <- apply(total.samples,1,mean)   
  all.sample.vars <- apply(total.samples,1,var) 
  
  par(mfrow=c(2,2))
  hist(my.samples[1,],col="gray",main="Distribuição de uma amostra",ylab="Frequência")
  hist(all.sample.sums,col="gray",main="Distribuição amostral da soma",ylab="Frequência")
  hist(all.sample.means,col="gray",main="Distribuição amostral da média",ylab="Frequência")
  hist(all.sample.vars,col="gray",main="Distribuição amostral da variância",ylab="Frequência")
  
}


#2b-idem ao 2a usando MC.
ex2b = function(){
    amostra_total <- runif(12000, min = 0, max = 1)
    for (i in 1:11) {
      amostras <- runif(12000, min = 0, max = 1)
      amostra_total <- amostra_total + amostras 
    }
    media <- mean(amostra_total)
    variancia <- sd(amostra_total)^2
    print(media)
    print(variancia)
    resultado <- rnorm(12000,media,variancia)
    hist(resultado)
}
  

#3- Usando MC, obtenha uma aproximaÃ§Ã£o empÃ­rica para a VA que representa a soma de 10 distribuiÃ§Ãµes triangulares Xi distribuidas com parÃ¢metros (mini = i; maxi = 20 + i; mprovi = 10 + i). Calcule a mÃ©dia, a variÃ¢ncia e plote um grÃ¡fico de sua funÃ§Ã£o de probabilidade.
ex3 = function(){
  media <- c()
  variancia <- c()
  for(n in 1:1000){
    u <- 0
    for(i in 1:10){
      u <- rtriangle(1000,i,20+i,10+1) + u
    }
    media <- c(media,mean(u))
    variancia <- c(variancia,var(u))
  }
  mean(media)
  mean(variancia)
  hist(u)
}

#4- Usando simulaÃ§Ã£o:
#4a- obtenha uma aproximaÃ§Ã£o empÃ­rica para a funÃ§Ã£o de probabilidade do produto de duas VAs Z = X âˆ— Y; X; Y âˆ¼ N(0; 1).
ex4a = function(n){
  z <- rnorm(n,0,1) * rnorm(n,0,1)
  hist(z)
}

#4b- idem para o quociente Z = X=Y .
ex4b = function(n){
  z <- rnorm(n,0,1) / rnorm(n,0,1)
  hist(z)  
}

#5-Obtenha um aproximaÃ§Ã£o empÃ­rica para a funÃ§Ã£o de probabilidade MÃ¡ximo(Xi)(i = 2; 5; 10) que representa a funÃ§Ã£o de probabilidade do mÃ¡ximo dentre i VAs cada uma delas seguindo uma Normal(0; 1).
ex5 = function(){
  len = 3000
  tam <- c(2,5,10)
  
  for(t in 1:3 ){
    conjunto <- c()
    for(l in 1:len){
      x1 <- -Inf
      for(t2 in 1:tam[t] )  x1 <- max(x1,rnorm(1,0,1))
      conjunto <- c(conjunto,x1)
    }
    hist(conjunto)
  }
}

#6-Obtenha uma aproximaÃ§Ã£o empÃ­rica para a funÃ§Ã£o Ï‡2(n) = Pn 1 Xi2 onde Xi âˆ¼ Normal(0; 1).
ex6 = function(){
  aprox <- function(n){
    z <- n *rnorm(1000,0,1) ^2
    hist(z)
  }
  for(h in 1:4)
    aprox(h)
}

#7-Obtenha uma aproximaÃ§Ã£ao empirica para a funÃ§Ã£o de probabilidade Z = eN(0;1)
ex7 = function(){
  z <- exp(rnorm(1000,0,1))
  hist(z)
}
```
