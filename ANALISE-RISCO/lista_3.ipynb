```r
library(triangle)
```

```r
construcao <- function(ns=1000){
    escv<-rtriangle(ns,75,92.5,82.5)
    fund<-rtriangle(ns,57.5,77.5,67.5)
    estr<-rtriangle(ns,430,472.5,445)
    telh<-rtriangle(ns,140,157.5,145)
    acab<-rtriangle(ns,72.5,107.5,92.5)
    cenarios <- cbind(escv,fund,estr,telh,acab)
    custo <- apply(cenarios, M=1, sum)
    hist(custo)
    plot(ecdf(custo))

    
}
construcao()
```

```r
media <- function(a) {
    (a[1]+a[2]+a[3])/3
}
desvpad <- function(a) {
    (a[1]^2 + a[2]^2 + a[3]^2 - a[1]*a[2] - a[2]*a[3] - a[1]*a[3])/18
}

escv1 <- c(75,92.5,82.5)
fund1 <- c(57.5,77.5,67.5)
estr1 <- c(430,472.5,445)
telh1 <- c(140,157.5,145)
acab1 <- c(72.5,107.5,92.5)

media_total <- media(escv1) + media(fund1) + media(estr1) + media(telh1) + media(acab1)
var_total <- desvpad(escv1) + desvpad(fund1) + desvpad(estr1) + desvpad(telh1) + desvpad(acab1)

media_total
dev_total = sqrt(var_total)
dev_total
```

```r
x1 <- media_total-2*dev_total
x2 <- media_total+2*dev_total
xplot <- x1:x2
plot(dnorm(x=xplot,media_total,dev_total))
```

```r
library(triangle)
custo <- function(prob = .35, ns =10000) {
    distancia <- (260 * (1-prob) + 290 * prob) * 1000 #### em metros
    n_tubos = ceiling(distancia / 8)
    tub <-    rtriangle(ns,725,790,740) * n_tubos ### dolar
    cavar <-  rtriangle(ns,12,25,16) * n_tubos   ### horas
    mdo <-    rtriangle(ns,17,23,18.5)  ### dolar/hora
    transp <- rtriangle(ns,6.1,7.4,6.6) * n_tubos ### dolar 
    sold <-   rtriangle(ns,4,5,4.5) * (n_tubos -1) ### horas
    filtr <-  rtriangle(ns,165000,188000,173000) ### dolar
    acab <-   rtriangle(ns,14000,17000,15000) * distancia/1000 ### dolar
    
c <- (tub + cavar * mdo + transp + sold* mdo + filtr + acab) / 1000000


}

hist(custo(0), col=rgb(.8,0,0,.3),breaks =100, xlim=c(35,57), probability = T)   ### red
hist(custo(1), col=rgb(0,0,.8,.3),breaks =100, xlim=c(35,57), probability = T, add=T)  ### blue
box()
```

```r
plot(ecdf(custo(0)), main="Probabilidade Cumulativa P(Custo<x)")
plot(ecdf(custo(.2)), add=T, col="red")
plot(ecdf(custo(.4)), add=T, col="blue")
plot(ecdf(custo(.6)), add=T, col="purple")
plot(ecdf(custo(.8)), add=T, col="cyan")
plot(ecdf(custo(1)), add=T, col="green")
```
