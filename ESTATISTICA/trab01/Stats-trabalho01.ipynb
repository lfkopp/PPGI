```python
years = []
for line in ref:
    match = re.match(r'.*?([1-2][0-9]{3})', line)
    if match is not None:
        years.append(int(match.group(1)))
print(years)
```

```python
labels = range(1993, 2017)
size = (15,7)
plt.figure(figsize=size, dpi=80)

n, y, _ = plt.hist(years,bins=labels, rwidth=0.9, facecolor='0507770', alpha=0.7)
plt.xticks(labels)
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.title('Occurrence of documents in "References" by year of publication')
plt.yticks([])
for i in range(len(y)-1):
    if n[i]: plt.text(y[i]+.5,n[i]-.6,int(n[i]),
                      {'ha': 'center', 'va': 'center'}, 
                      rotation=0)
plt.show()
```
