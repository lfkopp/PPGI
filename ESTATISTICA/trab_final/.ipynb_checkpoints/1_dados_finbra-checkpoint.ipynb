```python
import pandas as pd
import numpy as np
```

```python
dir = 'C:/_dados/_eleicoes/'
pd.set_option('display.max_columns', 500)
```

```python
finbra = pd.DataFrame({})
for y in range(2013,2018):
    f = pd.read_csv(dir + 'finbra-'+str(y)+'.csv', sep=';', encoding='windows-1252', skiprows=3, thousands='.', decimal=',', quotechar='"')
    f['ano'] = y
    finbra = pd.concat([finbra,f], ignore_index=True)
finbra.head()
```

```python
finbra2 = pd.DataFrame({})
for y in range(2003,2013):
    f = pd.read_excel(dir + 'finbra-'+str(y)+'.xlsx')
    f['ano'] = y
    finbra2 = pd.concat([finbra2,f], ignore_index=True)
    display(finbra2.tail())
finbra2.head()
```

```python
finbra3 = finbra2[['ano', 'UF', 'Total Geral da Despesa por Função', 'Populacao', 'MUNICIPIO', 'Despesas por Função', 'Comunicação Social', 'Comunicações', 'Cod Mun', 'CD_UF', 'CD_MUN' ]]
finbra3.fillna(0,inplace=True )
finbra3['C_S'] = finbra3['Comunicação Social'] + finbra3['Comunicações'] 
finbra3['CD_UF'] = max(finbra3['UF'] , finbra3['CD_UF'] )
finbra3.sort_values('C_S')
```
